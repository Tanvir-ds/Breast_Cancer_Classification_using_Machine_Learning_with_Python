```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset from Scikit-Learn 
from sklearn import datasets
sklearn_breast_cancer_data = datasets.load_breast_cancer()
sklearn_breast_cancer_data
```

```python
# Creating Pandas DataFrame
breast_cancer_data = pd.DataFrame(sklearn_breast_cancer_data['data'],columns=sklearn_breast_cancer_data['feature_names'])
breast_cancer_data
```

```python
# Adding target column to the DataFrame
breast_cancer_data['target'] = sklearn_breast_cancer_data['target']
breast_cancer_data
```

```python
# Finding missing values
breast_cancer_data.isnull().sum()
```

```python
# Finding duplicates value
breast_cancer_data.duplicated().sum()
```

```python
# Getting some information about the dataset
breast_cancer_data.info()
```

```python
# Statistical information about the dataset
breast_cancer_data.describe()
```

```python
# Number of entities of the target column
breast_cancer_data['target'].value_counts()
```

# 0 => Malignant
# 1 => Benigh

```python
# Visualization
sns.set()
sns.countplot(x='target',data=breast_cancer_data)
```

```python
# Extracting Independent and Dependent valriable
x = breast_cancer_data.drop(columns=['target'],axis=1)
y = breast_cancer_data['target']
```

```python
x
```

```python
y
```

```python
# Splitting the dataset into training and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(breast_cancer_data.shape,x_train.shape,x_test.shape)
```

# Training the Logistic Regression Model with train data

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```

# Model Evaluation

```python
# Predicting training dataset
predicted_train_data = model.predict(x_train)
predicted_train_data
```

```python
# Checking the accuray of training data 
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_train_data,y_train)
print("Training Data Accuracy:",score)
```

```python
# Predicting test dataset
predicted_test_data = model.predict(x_test)
predicted_test_data
```

```python
# Checking the accuray of test data 
from sklearn.metrics import accuracy_score
score = accuracy_score(predicted_test_data,y_test)
print("Test Data Accuracy:",score)
```

# Creating confusion matrix and visualizing

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(predicted_test_data,y_test)

sns.set()
sns.heatmap(cm,annot=True,fmt='.1f')
```
